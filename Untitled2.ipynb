```python
score = int(input('당신의 점수를 입력하세요'))
gr = ''

if score >= 90:
  gr = 'A'
elif score >= 80:
  gr = 'B'
elif score >= 70:
  gr = 'C'
elif score >= 60:
  gr = 'D'
else:
  gr = 'F'

if int(score%10) >= 7:
  gr = gr+'+'
elif int(score%10) >= 4:
  gr = gr+'0'
else:
  gr = gr+'-'

print(f'당신의 점수는 {score}이고 학점은 {gr}입니다.' )
```

```python
p = input('비밀번호 입력')
ch_count = 0
no_count = 0

for ch in p:
  if ch.isalpha():
    ch_count += 1
  elif ch.isnumberic():
    no_count += 1

if ch_count > 0 and no_count > 0:
  print('사용 가능한 비번입니다')
else:
  print('사용 불가한 비번입니다.')
```
