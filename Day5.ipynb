```python
def sum_all(start, end):
  output = 0
  for i in range(start, end+1):
    output += i
  return output

print('0에서 100까지의 합 :',sum_all(0,100))
print('0에서 100까지의 합 :',sum_all(0,1000))
print('0에서 100까지의 합 :',sum_all(50,100))
print('0에서 100까지의 합 :',sum_all(500,1000))
```

```python
#재귀 함수

def facto(n):
  output = 1
  for i in range(1,n+1):
    output *= i
  return output

print('3!:',facto(3))
print('3!:',facto(5))
print('3!:',facto(7))
```

```python
def facto(n):
  if n == 0:
    return 1
  else:
    return n*facto(n-1)
  

print('3!:',facto(3))
print('3!:',facto(5))
print('3!:',facto(7))
```

```python
# 피보나치 수열(토끼의 번식 속도, 바이러스)

def fibo(n):
  if n == 0:
    return 0
  elif n==1: return 1
  else:
    return fibo(n-1)+fibo(n-2)
  

output = int(input())



print('fibonacci(1):',fibo(1))
print('fibonacci(1):',fibo(2))
print('fibonacci(1):',fibo(3))
print('fibonacci(1):',fibo(4))
print('fibonacci(1):',fibo(5))
```

```python
def add_many(*args):
  result = 0

  for arg in args:
    result += arg
  return result

print(add_many(1,2,3))
print(add_many(1,2,3,4,5))
```

```python
def add_mul(choice,*args):
  if choice == 'add':
    result = 0
    for i in args:
      result += i
  elif choice == 'mul':
    result = 1
    for i in args:
      result = result*1
  return

print('add',1,2,3,4,5)
print('mul',1,2,3)
```
